### Explore hypnogram data

```python
from pandas import DataFrame

def read_hypno(hypnogram):
    hypno = list(map(int, eval(hypnogram)))
    hypno = np.array(hypno)
    return hypno

def quantile(nparray):
    return np.percentile(nparray, 0.75)

def sleep_duration(read_hypnogram):
    return len(read_hypnogram)

def deep_sleep_duration(read_hypnogram):
    # return np.count_nonzero(read_hypnogram >= 3)
    return np.count_nonzero(read_hypnogram == 4)

def phase_transitions(read_hypnogram):
    return np.count_nonzero(np.diff(read_hypnogram))

def x_shape(df):
    x = df.HYPNOGRAM
    xmes = x.apply(read_hypno)
    xmesmean = xmes.apply(np.mean)
    xmesstd = xmes.apply(np.std)
    xmesmax = xmes.apply(np.max)
    xmesquant= xmes.apply(quantile)
    xmeslen = xmes.apply(sleep_duration)
    xmesdslen = xmes.apply(deep_sleep_duration)
    xmestrans = xmes.apply(phase_transitions)
    hypno_df = DataFrame()
    hypno_df['duration'] = xmeslen
    hypno_df['deep_duration'] = xmesdslen
    hypno_df['deep_freq'] = xmesdslen / xmeslen
    hypno_df['transitions'] = xmestrans
    hypno_df['mean'] = xmesmean
    hypno_df['std'] = xmesstd
    hypno_df['max'] = xmesmax
    hypno_df['quant'] = xmesquant
    return hypno_df
```

```python
x = fulldf.HYPNOGRAM
y = fulldf['TARGET']
x_test = test_in.HYPNOGRAM    

xmes = x.apply(read_hypno)

xmesmean = xmes.apply(np.mean)
xmesstd = xmes.apply(np.std)
xmesmax = xmes.apply(np.max)
xmesquant= xmes.apply(quantile)

plt.subplot(4,1,1)
plt.scatter(y,xmesmean)
plt.subplot(4,1,2)
plt.scatter(y,xmesstd)
plt.subplot(4,1,3)
plt.scatter(y,xmesmax)
plt.subplot(4,1,4)
plt.scatter(y,xmesquant)
plt.show()

print("mean: "+ str(pearsonr(xmesmean,y)))
print("std: "+ str(pearsonr(xmesstd,y)))
print("max: "+ str(pearsonr(xmesmax,y)))
print("quant: "+ str(pearsonr(xmesquant,y)))
```

```python
plt.subplot(3,1,1)
plt.scatter(y,xmeslen)
plt.subplot(3,1,2)
plt.scatter(y,xmesdslen)
plt.subplot(3,1,3)
plt.scatter(y,xmestrans)
plt.show()
print("len: "+ str(pearsonr(xmeslen,y)))
print("dslen: "+ str(pearsonr(xmesdslen,y)))
print("trans: "+ str(pearsonr(xmestrans,y)))
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import learning_curve
```

```python
#nb_train = 550
nb_train = len(xmes)
svr.fit(hypno_df.iloc[0:nb_train,:], y.iloc[0:nb_train])
```

```python
y_pred = svr.predict(hypno_df.iloc[nb_train+1:,:])
```

```python
def score_function(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
score = score_function(y_pred, y.iloc[nb_train+1:])
print(score)
```
